## `print()` 参数

* **sep**：分隔符（默认空格）

  ```python
  print(1, 2, 3, sep='-')  # 1-2-3
  ```

* **end**：结尾（默认换行）

  ```python
  print("hi", end="")  # 不换行
  ```

* **file**：输出目标（默认屏幕）

  ```python
  print("log", file=open("a.txt","w"))
  ```
