## 题目
设计一个array类，这个类有三个操作

- insert(index, value) - 在index的地方insert value，如果index比array长，就放到array的结尾
- get(index) - 拿到index所在位置的 value
- fix() - 使得array是unmutable的比如insert(0, "a")


测试例子：

insert(0, "b")

insert(2, "c")

insert(2, "d")

你就得到一个array["b", "a", "d", "c"]

fix() 就这个不动了

get(0) -> b

get(1) -> a

```python
# 可以引⼊的库和版本相关请参考 “环境说明”
# Please refer to the "Environmental Notes" for the libraries and versions that can be introduced.

class Node:
    def __init__(self,val,isleft=False):
        self.cnt0 = 0
        self.cnt1 = 0
        self.val = val
        self.lef = None
        self.rig = None
        self.isleft= isleft   ##是左子树节点，还是右子数节点

    def addson(self,node,inleft):
        if inleft:
            self.lef = node
        else:
            self.rig = node
        

head = Node(val="a",isleft=True)

def insert(idx,c):
    now=head
    inleft=True    ##记录在左右
    last=head
    n=head.cnt0+head.cnt1+1    ##表示总个数
    while True:
        ##计算当前节点在第几个
        if now is None:
            now = Node(val=c,isleft=inleft)
            last.addson(now,inleft)
                
            break
        assert now != None
        if now.isleft:
            loc = now.cnt0+1
        else:
            loc = n-(now.cnt1-1)
        # print(f"now={now.val}, loc={loc}")
        
        last=now
        
        if idx<loc:
            now.cnt0 += 1
            now = now.lef
            inleft = True
        else:
            now.cnt1 += 1
            now = now.rig
            inleft = False


fix_vec=[]
def fix(node):
    if node is None:
        return
    fix(node.lef)
    fix_vec.append(node.val)
    fix(node.rig)


def get(idx):
    return fix_vec[idx]

if __name__ == "__main__":
    insert(0,"b")
    insert(2,"c")
    insert(2,"d")
    insert(5,"f")
    fix(head)
    print(fix_vec)

    print(get(0))
    print(get(1))
    
    
    
```
