
### 1.TCP 为什么要三次握手？


第一次握手是客户端请求建立连接，客户端向服务器发送 SYN 包，请求建立连接；

第二次握手时服务端说同意建立连接，服务器向客户端发送SYN+ACK；

第三次是客户端说我们开始建立连接。如果没有第3次握手，那么可能服务端会因为网络延迟收到一个请求，然后就建立连接，但是这时候客户端已经掉线了，从而浪费连接资源。

而第1和第2次握手证明客户端能发,能收。



### 2. 请你讲一下 哈希表的底层原理。 在 Python 的 dict 或 C++ 的 unordered_map 中，是如何实现键值映射的？ 当出现哈希冲突时，系统是怎么处理的？

> 哈希表的底层原理是“用哈希函数把 key 映射到数组下标”，通过这个数组实现 O(1) 级别的插入、删除和查找。
>
> 具体地：
>
> 1. **底层结构**：哈希表内部维护一个固定大小的数组（称为桶，bucket），每个 key 通过哈希函数计算出索引位置。
>
> 2. **冲突处理**：
>
>    * **Python 的 `dict`** 使用**开放寻址法（Open Addressing）**。当目标位置被占用时，它会沿着探测序列（如线性或伪随机）继续寻找下一个空位；查找时沿相同路径找回。
>    * **C++ 的 `unordered_map`** 使用**链地址法（Separate Chaining）**。每个桶存放一个链表（或 list/vector），当多个 key 落入同一桶时，就依次挂在链表后面。
>
> 3. **负载因子与扩容**：
>    当元素数量 / 桶数量（即负载因子）超过阈值（Python ~0.66，C++ 默认 1.0），会自动进行 **rehash**：扩容数组、重新计算每个 key 的哈希值、重新放入新桶中。
>
> 4. **复杂度**：
>
>    * 平均时间复杂度：O(1)
>    * 最坏情况：O(N)（当所有元素冲突到同一个桶）
>
> 因此，哈希表通过空间换时间，利用数组的随机访问特性，实现高效映射。




### 3.TCP 四次挥手（断开连接） 为什么不是三次，而是四次？


> TCP 是**全双工通信**，也就是说，客户端和服务器之间的数据流是**两个独立的方向**。
> 因此，断开连接时，每个方向都要独立地“请求关闭（FIN）→ 确认关闭（ACK）”。
>
> **1️⃣ 第一次挥手（FIN）：**
> 客户端说「我不再发数据了」，进入 FIN_WAIT_1。
>
> **2️⃣ 第二次挥手（ACK）：**
> 服务器确认「我收到了你的关闭请求」，但它可能还有数据没发完。
>
> **3️⃣ 第三次挥手（FIN）：**
> 服务器发完数据后，也发 FIN，表示「我也发完了」。
>
> **4️⃣ 第四次挥手（ACK）：**
> 客户端回复 ACK，确认服务器的关闭，连接彻底断开。
>
> 三次不够的原因是：
> 服务器不能在“确认客户端关闭”的同时马上关闭自己（可能还有数据没发完），
> 所以**必须再多一次独立的 FIN**。
