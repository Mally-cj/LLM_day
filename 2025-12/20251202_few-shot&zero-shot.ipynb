## Few-shot CoT
**特点**：你先给 1–3 个带思维链的示例，再给目标问题，模型模仿结构输出 CoT。

输入：
```
【示例 1】
问题：一辆公交车上有 30 个人，到站后下去 12 人，又上来 5 人。车上还有多少人？
思考：原来 30 人，下去 12 人后剩 18 人，上来 5 人变成 23 人。
答案：23

【示例 2】
问题：小李有 50 元，买书花了 28 元，又买水花了 3 元，还剩多少钱？
思考：50 - 28 = 22，22 - 3 = 19。
答案：19

【要解决的问题】
问题：小明有 15 个苹果，送给同学 7 个，又买了 4 个，他现在有多少个？
```

输出：
```
思考：15 - 7 = 8，8 + 4 = 12。
答案：12
```

## Zero-shot CoT
特点：不给示例，只靠一条指令让模型开启“推理步骤模式”


输入：
```
小明有 15 个苹果，送给同学 7 个，又买了 4 个。
他现在有多少个苹果？
让我们一步一步思考。
```

输出：
```
首先，小明原来有 15 个苹果。
送给同学 7 个后，还剩 15 - 7 = 8 个。
后来又买了 4 个，所以现在有 8 + 4 = 12 个。
答案是：12。
```
